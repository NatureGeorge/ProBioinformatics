```r
library(MASS)
g1 <- split(Cars93$MPG.city, Cars93$Origin)
g1
cat(median(g1[[1]]),median(g1[[2]]))
```

```r
g2 <- unstack(data.frame(Cars93$MPG.city, Cars93$Origin))
g2
cat(median(g2[[1]]),median(g2[[2]]))
```

```r
g_3_1 <- Cars93[which(Cars93$Origin=='USA'),]['MPG.city']
g_3_2 <- Cars93[which(Cars93$Origin=='non-USA'),]['MPG.city']
cat(median(g_3_1[[1]]), median(g_3_2[[1]]))
```

```r
g4 <- by(Cars93$MPG.city, Cars93$Origin, mean)
g4
cat("============================================================\n")
g4_2 <- by(Cars93$MPG.city, Cars93$Origin, summary)
g4_2
cat("============================================================\n")
cat(g4_2$USA['Mean'], g4_2[['non-USA']]['Mean'])
```

```r
g5 <- tapply(Cars93$MPG.city, Cars93$Origin, mean)
g5
```
