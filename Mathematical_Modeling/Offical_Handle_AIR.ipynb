```python
def plotAIR(df, air, tp=("HOUR", "DAY", "COUNT"), figsize=(20,10)):
    new_count_mat = df[df['AIR']==air].pivot(*tp)
    fig = plt.figure(figsize=figsize)
    return sns.heatmap(new_count_mat.fillna(0))
```

```python
def plotMonthCount(df, air='NAY', time_col='DAY', data_col='MONTH_COUNT',figsize=(30,5)):
    vis_df = df[df['AIR']==air]
    fig = plt.figure(figsize=figsize)
    plt.plot(pd.to_datetime(vis_df[time_col]), vis_df[data_col])
    plt.xticks(rotation=45)
```
