# Bioinformatics Algorithm

## 过程化考核5: Biological Network

> Last Modified Time 2019-11-25 15:18 By ZeFengZhu

### 问题描述

列举常见生物网络，并说明网络中的点和边分别是什么

### 基础概念


#### Reference

> <https://www.ebi.ac.uk/training/online/course/network-analysis-protein-interaction-data-introduction/networks-cell-biology-summary-0>

一些最常见的生物网络类型包括：
1. 蛋白质-蛋白质相互作用网络(Protein-protein interaction networks)
2. 代谢网络(Metabolic networks)
3. 遗传相互作用网络(Genetic interaction networks)
4. 基因/转录监管网络(Gene / transcriptional regulatory networks)
5. 细胞信号网络(Cell signalling networks)

![fig](https://www.ebi.ac.uk/training/online/sites/ebi.ac.uk.training.online/files/resize/Networks_cellnetworksall-700x370.png)

> 不同类型的数据还将在连接性、复杂性和结构方面产生不同的一般网络特征，其中边缘和节点可能传达多层信息

#### Protein-protein interaction networks

![fig](https://www.ebi.ac.uk/training/online/sites/ebi.ac.uk.training.online/files/slides/Fig_protein_networks_new.png)


* 表示蛋白质之间的物理关系。它们对于细胞中发生的几乎每一个过程都是核心
* Node: 蛋白质表示为由无定向边链接的节点
* edges:蛋白质间的相互作用为无向边

#### Metabolic networks

![fig](https://www.ebi.ac.uk/training/online/sites/ebi.ac.uk.training.online/files/slides/FIg_metabolism_network_new.png)

* 代表生物化学反应，使生物体生长、繁殖、对环境作出反应并维持其结构
* Node:代谢物和酶扮演节点的角色
* edges:描述其转化的反应表示为有向边
* 边可以表示特定反应的代谢流动或调节作用的方向
 
#### Genetic interaction networks

![fig](https://www.ebi.ac.uk/training/online/sites/ebi.ac.uk.training.online/files/slides/Networks_genetic%20interaction_new.png)

* 遗传相互作用是协同现象，其中由两个或两个以上基因的同步突变产生的表型与添加单个突变的影响所产生的表型有显著差异
* 代表不同基因之间的功能关系，而不是物理基因
* Node:基因表示为节点
* edges:其关系为边， 根据交互背后的证据类型，可以推断边的方向性

#### Gene / transcriptional regulatory networks

![fig](https://www.ebi.ac.uk/training/online/sites/ebi.ac.uk.training.online/files/slides/Fig_gene_reg_network_new.png)

* 表示如何调控基因表达
* Node:基因和转录因子表示为节点
* edges: 而它们之间的关系由不同类型的有向边表示
* 调控RNA和其他机制也可以构成这类网络的一部分
* 通常通过代表基因调控共识知识的数据库生成，例如`Reactome`, `KEGG`，尽管大型实验数据集越来越多

#### Cell signalling networks

![fig](https://www.ebi.ac.uk/training/online/sites/ebi.ac.uk.training.online/files/slides/Fig_cell_signalling_network_new.png)

* 细胞信号是控制细胞活动的通信系统
* 信号通路表示有序的事件序列，并模拟单元内的信息流
* 基因调节网络可被视为细胞信令网络的子类型，侧重于特定信号事件，该事件通常是信令级联的最后阶段
* Node: 通路中的元素（例如蛋白质、核酸、代谢物）表示为节点，
* edges: 信息流由有向边表示
* 系统地由两种类型的资源表示：
    * Pathway数据库（也称为"过程描述"资源），例如`Reactome`, `KEGG`和`Wikipathways`。其目的是为目前关于细胞信号通路的科学共识提供正式的表示。它们由人工固化生成(manually curated)，以反应的形式组织信息，基质和产品受催化器作用的影响。此信息必须根据特定规则进行转换，才能表示为网络。在此过程中，可能会出现一些信息丢失
    * Reaction Network数据库（也称为"活动流"资源），如`Signor`、`SignaLink`和`SPIKE`，这些旨在捕获细胞信令中的已知二进制关系，如活化、磷酸化等。它们通常是手动策划的，但并非总是如此。与通路数据库相比，它们已经是数学意义上的图形，不需要变换才能表示为网络。
